```python
import numpy as np
import pandas as pd
import faiss

# ✅ Load books dataset
books_df = pd.read_pickle("books_with_expanded_interest.pkl")

# ✅ Load FAISS indices
index_st = faiss.read_index("faiss_st.index")  # SentenceTransformer FAISS index
index_t5 = faiss.read_index("faiss_t5.index")  # T5 FAISS index

# ✅ Load precomputed embeddings
st_matrix = np.load("st_embeddings.npy")  # SentenceTransformer embeddings
t5_matrix = np.load("t5_matrix.npy")  # T5 embeddings

# ✅ Ensure FAISS index and books_df match
assert st_matrix.shape[0] == t5_matrix.shape[0] == books_df.shape[0], "⚠️ Mismatch between embeddings and books_df!"


def recommend_books(user_reviews, top_k=5):
    """
    Recommend books based on expanded interest embeddings using T5 and SentenceTransformer.
    Uses precomputed FAISS indices instead of recomputing embeddings.
    """
    # ✅ Retrieve expanded interest safely
    matched_rows = books_df[books_df["review/text"].str.contains(user_reviews, case=False, na=False)]
    
    if not matched_rows.empty:
        expanded_interest = matched_rows["expanded_interest"].iloc[0]  # ✅ Take first match
    else:
        print("⚠️ No exact match found for user review. Using default interest.")
        expanded_interest = "science fiction, space adventure, futuristic exploration"  # Default

    print("\n🔹 **Expanded Interest:**", expanded_interest)

    # ✅ Load precomputed SentenceTransformer embeddings
    query_embedding_st = np.load("st_embeddings.npy")[0].astype("float32").reshape(1, -1)

    # ✅ Load precomputed T5 embeddings
    query_embedding_t5 = np.load("t5_matrix.npy")[0].astype("float32").reshape(1, -1)

    # ✅ Search in FAISS index (ST-based)
    distances_st, indices_st = index_st.search(query_embedding_st, top_k)

    # ✅ Search in FAISS index (T5-based)
    distances_t5, indices_t5 = index_t5.search(query_embedding_t5, top_k)

    # ✅ Retrieve recommended books
    recommended_st = books_df.iloc[indices_st[0]][["Title", "review/summary"]].copy()
    recommended_t5 = books_df.iloc[indices_t5[0]][["Title", "review/summary"]].copy()

    return expanded_interest, recommended_st, recommended_t5

# =========================
# **TEST EXAMPLE**
# =========================
user_reviews = "I love epic sci-fi novels with deep world-building."

expanded_interest, recommended_st, recommended_t5 = recommend_books(user_reviews)

print("\n🔹 **Expanded Interest:**", expanded_interest)
print("\n🔹 **Recommended Books (SentenceTransformer-based):**")
print(recommended_st)

print("\n🔹 **Recommended Books (T5-based):**")
print(recommended_t5)
```
